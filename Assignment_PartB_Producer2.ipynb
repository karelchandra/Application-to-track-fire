```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv
```

```python
def publish_message(producer_instance, topic_name, data):
    try:
        value_bytes = bytes(data, encoding='utf-8')
        producer_instance.send(topic_name, value=value_bytes)
        producer_instance.flush()
        print('Message published successfully. Data: ' + data)
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
```

```python
# Establish Connection with kafka
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
```

```python
if __name__ == '__main__':
   
    topic = 'EventProducer01'
    
    print('Publishing records..')
    producer = connect_kafka_producer()
    # Find the latest date
    latest_date = dt.datetime(2021,12,31)
    new_date = latest_date + dt.timedelta(days=1)
    
    # Open folder of data from csv
    with open('hotspot_AQUA_streaming.csv') as hotspot_AQUA:
        hotspotA = csv.DictReader(hotspot_AQUA)
        optionHA = [i for i in hotspotA]
        # Run for 100 times
        for e in range(100):
            # Random the data everytime
            chosen_row = random.choice(optionHA)
            # Add new columns into the data
            chosen_row['date'] = new_date.strftime("%Y-%m-%d %H:%M:%S")
            chosen_row['producer_num'] = 'Producer02'
            data = str(chosen_row)
            # Push to kafka
            publish_message(producer, topic, data)
            # Wait for 2 seconds before going again
            sleep(2)
            new_date = new_date + dt.timedelta(hours=4.8)
```
